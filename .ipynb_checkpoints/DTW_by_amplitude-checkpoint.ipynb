## Define function

```python
def meanSquaredError(a, b):
    '''
    a, b (float): answer to the meanSquaredError. 
    '''
    return ((a-b)**2).mean()

def firstIndex(x):
    '''
    x (array): answer to First dimension of the array. 
    '''
    return x[0]

def secondIndex(x):
    '''
    x (array): answer to Second dimension of the array. 
    '''
    return x[1]

def minVal(v1, v2, v3):
    '''
    v1, v2, v3 (tuple): answer to the minimum meanSquaredError and the index. 
    '''
    if firstIndex(v1) <= min(firstIndex(v2), firstIndex(v3)):
        return v1, 0
    elif firstIndex(v2) <= firstIndex(v3):
        return v2, 1
    else:
        return v3, 2 

def calc_dtw(short_amp, long_amp):
    # get the frame of the audio
    S = len(short_amp)
    L = len(long_amp)

    # Array initialization (Rows:S, Colums:L)
    m = [[0 for j in range(L)] for i in range(S)]
    
    # Insert a tuple consisting of first frame meanSquaredError and (-1, -1) into the first cell
    m[0][0] = (meanSquaredError(short_amp[0],long_amp[0]), (-1,-1))
    
    # Insert a tuple of total meanSquaredError in the previous cell and (i-1, 0) into the first Colum Cells
    for i in range(1,S):
        m[i][0] = (m[i-1][0][0] + meanSquaredError(short_amp[i], long_amp[0]), (i-1,0))
    # Insert a tuple of total meanSquaredError in the previous cell and (0, j-1) into the first Row Cells
    for j in range(1,L):
        m[0][j] = (m[0][j-1][0] + meanSquaredError(short_amp[0], long_amp[j]), (0,j-1))
    
    # Insert a tuple into Empty Cells
    for i in range(1,S):
        for j in range(1,L):
            # Add own meanSquaredError to the minimum meanSquaredError on the left, top, and top left
            minimum, index = minVal(m[i-1][j], m[i][j-1], m[i-1][j-1])
            indexes = [(i-1,j), (i,j-1), (i-1,j-1)]
            m[i][j] = (firstIndex(minimum) + meanSquaredError(short_amp[i], long_amp[j]), indexes[index])

    return m

def correnspondence(m):
    path = []
    path.append([len(m)-1, len(m[0])-1])
    while True:
        path.append(m[path[-1][0]][path[-1][1]][1])
        if path[-1]==(0,0):
            break
    path = np.array(path)
    return path
```

## check by sin wave

```python
T = 100
t = .4

A = np.sin(np.array(range(T))/10)
B = np.sin((np.array(range(T))/10 + t*np.pi))

plt.plot(A)
plt.plot(B)
plt.show()
```

```python
dist = calc_dtw(A, B)
print("wave distance: ", dist[-1][-1][0])
```

```python
path = correnspondence(dist)
for line in path:
    plt.plot(line, [A[line[0]], B[line[1]]], linewidth=0.2, c="gray")
plt.plot(A, marker='o')
plt.plot(B, marker='o')
plt.show()
```

## check by audio data

```python
dist_amp = calc_dtw(data_s, data_l)
print("wave distance: ", dist_amp[-1][-1][0])
```

```python
path_amp = correnspondence(dist_amp)
for line in path_amp:
    plt.plot(line, [data_s[line[0]], data_l[line[1]]], linewidth=0.2, c="gray")
plt.plot(data_l)
plt.plot(data_s)
plt.show()
```

```python
short_list = []
long_list = []
for line in path_amp:
    short_list.append(line[0])
    long_list.append(line[1])

plt.title('correnspondence frame number')
plt.plot(long_list, short_list)
plt.xlabel('long amptitude')
plt.ylabel('short amptitude')
plt.show()
```

```python
dtw_data_s = []
for short_index in short_list:
    dtw_data_s.append(data_s[short_index])
plt.title('DTW short amptitude')
plt.plot(dtw_data_s)
plt.show()

dtw_data_l = []
for long_index in long_list:
    dtw_data_l.append(data_l[long_index])
plt.title('DTW long amptitude')
plt.plot(dtw_data_l)
plt.show()
```
